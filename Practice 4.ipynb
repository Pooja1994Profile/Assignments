```python
#1. create a sample JSON object from titanic train dataset
import json
import csv
csvFilePath = "C:/Users/POOJA/titanic.csv"
jsonFilePath= "C:/Users/POOJA/json_file_name.json"
data = []
with open(csvFilePath) as csvFile:
    csvReader=csv.DictReader(csvFile)
    field = csvReader.fieldnames
    for row in csvReader:
        data.extend([{field[i]:row[field[i]] for i in range(len(field))}])
with open(jsonFilePath,'w') as jsonFile:
    jsonFile.write(json.dumps(data,indent=4))
```
